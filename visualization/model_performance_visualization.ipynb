```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_model = pd.read_csv("./data/model_perf.csv")
df_model = df_model[df_model["model"]!="LR"]
for year in [2006, 2061]:
    df = pd.read_parquet(f'../evaluation/{year}_LR.parquet.gzip')
    dfappend = df.groupby(['model', 'surf', 'loc']).mean().reset_index()
    dfappend = dfappend.rename(columns={'rmse': 'test-rmse', 'r2': 'test-r2', 'mae': 'test-mae'})
    dfappend['year'] = year
    df_model = pd.concat([df_model,dfappend])
df_model = df_model.replace("xgboost", "XGBoost")
df_model
```

```python
df_model = df_model[df_model["model"]!="lgbm"]
print('best R2: ', round(df_model['test-r2'].max(),2))
print("lowest R2: ", round(df_model['test-r2'].min(),2))
print("lowest RMSE: ", round(df_model['test-rmse'].min(),2))
print("lowest MAE: ", round(df_model['test-mae'].min(),2))
```

## model performance

## R2

```python
#df_model = pd.read_csv("./data/model_perf.csv")
df_model = df_model[df_model["model"]!="lgbm"]
# df_06 = df_model[df_model["year"]==2006]

# change the ylim for visualization
sns.set(rc={'figure.figsize':(12,3), "xtick.bottom":False}, font_scale=1.7, style="ticks")
g = sns.catplot(x="loc", y="test-r2", hue="surf", col="model", row='year', kind="bar",data=df_model,
                #palette=["skyblue","orange", ],
                edgecolor=".2", linewidth=1,
                #palette="Set1"
                legend=False
                )
i =0
for ax in g.axes.flat:
    sns.despine(left=False, right=False, top=False, bottom=False, ax=ax)
    if i == 0 or i == 3:
        ax.tick_params(axis='x', which='both', length=0, width=0, direction='out')  # x 轴刻度
        ax.tick_params(axis='y', which='both', length=8, width=2, direction='out')  # y 轴刻度
    else:
        ax.tick_params(axis='x', which='both', length=0, width=0, direction='out')
        ax.tick_params(axis='y', which='both', length=0, width=0, direction='out', labelleft=False)
    
    if i in [0, 3]:
        baseline = ax.containers[0][0].get_height()
        print('baseline of YY FLAML: ', baseline)
    print(f'ax{i}: ')
    #if i != 0 and i <= 5:
    #if i in [0,3]:
    if i in [0, 3]:
        labels_baseline = [g.axes.flat[i].containers[j0][j1].get_height() for j0 in range(0, 2) for j1 in range(0, 2)]
    print(f'baseline of ax{i}: ', labels_baseline)
    print(f'values of ax{i}: ', [g.axes.flat[i].containers[j0][j1].get_height() for j0 in range(0, 2) for j1 in range(0, 2)])
    j = 0
    for c in ax.containers:
        labels = [h.get_height() for h in c]
        labels = [f'{round(l,2)}\n(+{round(l-baseline,2)})' if round(l-baseline,2)!=0 else f'{round(l,2)}' for l in labels]
        #if i not in [0, 3]:
        #    labels = [f'+{round(h.get_height()-labels_baseline[m],2)}' for h,m in zip(c,[j,j+1])]
        print(labels)
        ax.bar_label(c, labels=labels, label_type='edge', padding=3)
        j += 2
    i += 1
g.set_ylabels("Test $R^2$")
g.set(ylim=0.75)
g.set(xlabel="Including locations")

fig = g.fig
handles, labels = g.axes[0, 0].get_legend_handles_labels()
legend1 = fig.legend(handles, labels, loc='lower center',
                     bbox_to_anchor=(0.5, -.08), 
                     ncol=2, title="Including surface parameters", 
                     frameon=False)
titles = ['a', 'b', 'c', 'd', 'e', 'f']  
for ax, title in zip(g.axes.flat, titles):
    ax.set_title(f"({title}) {ax.get_title()}")  
plt.savefig('./paper_figs/model_performance_r2.pdf', dpi=188, bbox_inches='tight')
plt.show()
```

## model performance under climate change

```python
df = pd.read_csv("./data/climate_perf.csv")
df = df[df["model"]!="LR"]

df1 = pd.read_parquet(f'../evaluation/climate_LR.parquet.gzip')
dfappend = df1.groupby(['model', 'surf', 'loc']).mean().reset_index()
df2 = pd.read_parquet(f'../evaluation/2061_LR.parquet.gzip')
dfappend2 = df2.groupby(['model', 'surf', 'loc']).mean().reset_index()
dfappend2 = dfappend2.rename(columns={'rmse': 'baseline-rmse', 'r2': 'baseline-r2', 'mae': 'baseline-mae'})
dfappend = dfappend.merge(dfappend2, on=['model', 'surf', 'loc'])
df = pd.concat([df,dfappend])
df = df.replace("xgboost", "XGBoost")
df
```
